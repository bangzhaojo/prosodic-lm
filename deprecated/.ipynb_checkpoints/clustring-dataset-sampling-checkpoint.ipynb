### Sampling

```python
date_counts = df['date'].value_counts(normalize=True)
```

```python
sample_counts = (date_counts * 10000).round().astype(int)
```

```python
sample_counts
```

```python
df['date']
```

```python
# Create an empty DataFrame to hold the stratified sample
stratified_sample_df = pd.DataFrame()

# Loop through each date and get the appropriate number of samples
for date, count in sample_counts.items():
    date_df = df[df['date'] == date].sample(n=count, random_state=1)
    stratified_sample_df = pd.concat([stratified_sample_df, date_df])

# If rounding made us sample too few or too many, adjust by randomly dropping/adding some samples
diff = 10000 - stratified_sample_df.shape[0]
if diff > 0:
    extra_samples = df.loc[~df.index.isin(stratified_sample_df.index)].sample(n=diff, random_state=1)
    stratified_sample_df = pd.concat([stratified_sample_df, extra_samples])
elif diff < 0:
    stratified_sample_df = stratified_sample_df.sample(n=10000, random_state=1)
```

```python
def process_and_write_json_objects(file_path, output_json_path):
    with open(file_path, 'r') as file, open(output_json_path, 'w') as output_file:
        for i, line in enumerate(file):
            if i in sample_idx:
                json_object = json.loads(line)
                output_file.write(json.dumps(json_object) + '\n')
            if i % 100000 == 0:
                print(f"Processed {i} lines..")
```

```python
def adding_prosodic_to_jsonl(file_path, output_json_path):
    with open(file_path, 'r') as file, open(output_json_path, 'w') as output_file:
        for i, line in enumerate(file):
            json_object = json.loads(line)

            prosodic_path = "/shared/3/projects/benlitterer/podcastData/prosodyMerged/floydMonth" + json_object['potentialOutPath']
            prosodic_feature = pd.read_csv(prosodic_path)
            if 'Unnamed: 0' in prosodic_feature.columns:
                prosodic_feature = prosodic_feature.drop(columns=['Unnamed: 0'])
            prosodic_feature = prosodic_feature.drop(columns=['index'])

            # Convert the DataFrame into a list of dictionaries (one per row)
            prosodic_data_list = prosodic_feature.to_dict('list')
            json_object.update(prosodic_data_list)

            output_file.write(json.dumps(json_object) + '\n')
                
            if i % 1000 == 0:
                print(f"Processed {i} lines..")
```

```python
def creating_prosodic_csv(file_path, output_json_path):
    all_prosodic = pd.DataFrame()
    with open(file_path, 'r') as file:
        for i, line in enumerate(file):
            json_object = json.loads(line)

            # Construct the full path to the CSV file containing prosodic features
            prosodic_path = "/shared/3/projects/benlitterer/podcastData/prosodyMerged/floydMonth" + json_object['potentialOutPath']
            prosodic_feature = pd.read_csv(prosodic_path)
            if 'Unnamed: 0' in prosodic_feature.columns:
                prosodic_feature = prosodic_feature.drop(columns=['Unnamed: 0'])
            prosodic_feature['potentialOutPath'] = json_object['potentialOutPath']

            # Concatenate the current DataFrame to the cumulative one
            all_prosodic = pd.concat([all_prosodic, prosodic_feature], ignore_index=True)
            
            if (i+1) % 100 == 0:
                print(f"Processed {i+1} lines..")
            if (i+1) % 1000 == 0:
                print(f"Processed {i+1} lines..")
                c = (i+1) // 1000
                all_prosodic.to_csv(output_json_path.split('.')[0]+str(c)+'.csv', index=False)
                all_prosodic = pd.DataFrame()

    # Write the final DataFrame to a CSV file
    all_prosodic.to_csv(output_json_path, index=False)
```

```python
from scipy.optimize import linear_sum_assignment
from scipy.spatial.distance import cdist

def match_centroids(centroids1, centroids2):
    # Compute the pairwise distance matrix
    distance_matrix = cdist(centroids1, centroids2, metric='euclidean')
    # Solve the assignment problem
    row_ind, col_ind = linear_sum_assignment(distance_matrix)
    # Return the total distance for the best assignment
    total_distance = distance_matrix[row_ind, col_ind].sum()
    return total_distance

# Example usage
total_dist = match_centroids(centroids_1k, centroids_2k)
print("Total distance between matched centroids:", total_dist)
```

```python
# Example usage
total_dist = match_centroids(centroids_1k, centroids_5k)
print("Total distance between matched centroids:", total_dist)
```

```python
inertias = []
k_values = [50, 100, 200, 500, 1000]

for k in k_values:
    model = KMeans(n_clusters=k, random_state=0)
    model.fit(X)
    inertias.append(model.inertia_)

# Plotting the elbow graph
plt.plot(k_values, inertias, '-o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia (WCSS)')
plt.xticks(k_values)
plt.show()
```
